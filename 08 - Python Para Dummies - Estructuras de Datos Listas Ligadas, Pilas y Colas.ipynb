```python
class ListaLigada:

  def __init__(self):
    self.cabeza = None
    self.tamano = 0

  def unir(self, dato):
    nodo = Nodo(dato, None)

    if self.cabeza == None:
      self.cabeza = nodo
    else:
      posicion_actual = self.cabeza

      while posicion_actual.siguiente:
        posicion_actual = posicion_actual.siguiente

      posicion_actual.siguiente = nodo

    self.tamano += 1
```

```python
class Pila:

  def __init__(self):
    self.top = None
    self.tamano = 0

  def agregar(self, dato):
    nodo = Nodo(dato, None)

    if self.top:
      nodo.siguiente = self.top
      self.top = nodo
    else:
      self.top = nodo

    self.tamano += 1
```

```python
heroes = Pila()
```

```python
heroes.agregar("superman")
```

```python
heroes.agregar("batman")
heroes.agregar("iron man")
```

```python
heroes.top.siguiente.siguiente.dato
```
