#Adivine el numero

```python
import random

numero_pensado = random.randint(0, 50)
numero_del_usuario = -1
vidas = 3

while numero_del_usuario != numero_pensado and vidas > 0:

  numero_del_usuario = int(input("Por favor ingrese un numero entre 0 y 50 "))
  vidas -= 1
  if numero_del_usuario > numero_pensado:
    print("Ingrese un numero menor")
  else:
    print("Ingrese un numero mayor")  

if vidas == 0:
  print("perdio")
else:
  print("gano")
```

#Ahorcado

```python
palabra_secreta = "python"
vidas = 3
letras_usadas = []
letras_correctas = []
espacios_en_blanco = "-" * len(palabra_secreta)
jugando = True

while jugando:

  print(espacios_en_blanco)
  letra = input("Por favor ingrese una letra ")

  if letra in palabra_secreta:
    print("la letra esta")
    letras_correctas.append(letra)

    for i in range(len(palabra_secreta)):
      if palabra_secreta[i] in letras_correctas:
        espacios_en_blanco = espacios_en_blanco[:i] + palabra_secreta[i] + espacios_en_blanco[i + 1:]

    if not "-" in espacios_en_blanco:
      print("gano")
      jugando = False

  else:
    vidas -= 1
    letras_usadas.append(letra)
    print("La letra no esta")
    if vidas == 0:
      print("perdio")
      jugando = False
```

#Sudoku

```python
sudoku = []

for i in range(1, 10):
  fila = [0] * 9
  sudoku.append(fila)

for i in range(0, 9):
  print(sudoku[i])
```

```python
class Matriz:

  def __init__(self, filas = 9, columnas = 9):
    self.filas = filas
    self.columnas = columnas
    self.posicion = []

    for fila in range(self.filas):
      self.posicion.append([0] * self.columnas)
  
  def llenar_valores_aleatorios(self):
    for fila in range(self.filas):
      for columna in range(self.columnas):
        self.posicion[fila][columna] = random.randint(1, 9)

  def agregar_valor(self, fila, columna, valor):
    if valor not in self.posicion[fila]:
      self.posicion[fila][columna] = valor    

  def ver_tablero(self):
    for fila in range(self.filas):
      print(self.posicion[fila])
```

```python
sudoku = Matriz()
```

```python
sudoku.llenar_valores_aleatorios()
sudoku.ver_tablero()
```

```python
sudoku.agregar_valor(0, 0, 7)
sudoku.ver_tablero()
```
