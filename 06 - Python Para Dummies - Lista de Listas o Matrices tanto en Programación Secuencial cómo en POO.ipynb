# Creacion de Matrices

### Una matriz es una lista de lista, por lo tanto todos los metodos y logica que se aplica en una lista, aplican tambien para matrices. len, pop, append, insert, entre otros funcionan tambien con matrices

```python
#Se crea una lista de listas y se almacena en una variable llamada matriz
matriz = [[1, 5, 7, 8], [2, 5, 9, 8], [8, 7, 9, 2]]
print(matriz)
```

# Acceder a los valores de una matriz

```python
#Se accede a una fila de la matriz usando el corchete cuadrado [] y dentro la posicion de la fila deseada. 
#Recuerda que se cuenta desde cero, osea que la posicion 1 seria la fila numero dos
print(matriz[1])
```

```python
#Se accede a un elemento usando dos corchetes cuadrados [][] con la posicion de la fila y la columna deseada
#Recuerda que se cuenta desde cero, osea que la posicion 1 seria la fila numero dos y la posicion 2 seria la columa numero tres
print(matriz[1][2])
```

```python
#Para cambiar el valor de un elemento de la matriz, se debe igualar la variable con las posiciones deseadas con un valor deseado
matriz[2][0] = 1000
print(matriz)
```

# Creación de una Matriz usando Ciclos

```python
#Variables para definir el tamaño de la matriz deseada
columnas = 3
filas = 3

#Lista vacia en donde se almacenará la matriz
nueva_matriz = []

#For que genera una lista de tamaño "columnas" y une esta lista a la matriz con append
for fila in range(filas):
  nueva_matriz.append([0] * columnas)

#Se imnprime la matriz nula creada previamente
print(nueva_matriz)

#Ciclo que permite que el usuario ingrese valores para cada posicion de la matriz nula creada previamente
for fila in range(filas):
  for columna in range(columnas):
    nueva_matriz[fila][columna] = float(input())

print(nueva_matriz)
```

#Recorrido de matrices para sumar o multiplicar elementos

```python
suma = 0
multiplicacion = 1

for fila in range(filas):
  for columna in range(columnas):
    suma = suma + nueva_matriz[fila][columna]
    multiplicacion = multiplicacion * nueva_matriz[fila][columna]

print(suma)
print(multiplicacion)
```
