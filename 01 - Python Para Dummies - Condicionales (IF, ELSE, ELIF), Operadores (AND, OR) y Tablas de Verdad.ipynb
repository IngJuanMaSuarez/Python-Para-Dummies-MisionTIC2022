### Programa Para Determinar si el año es Bisiesto

```python
anio = 2018

if anio % 4 == 0:
  if anio % 100 != 0 or anio % 400 == 0:
    print("El anio es bisiesto")
  else:
    print("El anio no es bisiesto")
else:
  print("El anio no es bisiesto")
```

```python
anio = 2014

if anio % 4 == 0 and (anio % 100 != 0 or anio % 400 == 0):
  print("El anio es bisiesto")
else:
  print("El anio no es bisiesto")
```

###Numeros Primos

```python
numero_a_revisar = 9

for numero in range(2, numero_a_revisar):
  if numero_a_revisar % numero == 0:
    print("No Primo")
  else:
    print("Primo")
```
